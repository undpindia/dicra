## Preamble

* First, from prior knowledge as well as the_info class_, it is important to note that several of the features/predictors from their definitions are not good variable to determine the objective of the task - if the fires were agricultural or not. They include:  state instrument used, that satellite track path, satellite scan (instrument scan),  version, type, year, FID_2, gridcode, count_grid, area, OBJECTID, FID_1, geometry
* From all investigation and research, agricultural planting and harvesting varies depending on the crop to be planted. Generally, most planting and harvesting happen in the between April and July and between September and mid December.


```python
fire_lulc_df = fire_lulc.drop(labels=['scan', 'track', 'satellite', 'instrument', 'latitude', 'longitude', 'version', 'type', 'year', 'FID_2', 'gridcode', 'COUNT_grid', 'area', 'OBJECTID', 'FID_1', 'geometry'], axis =1)
```

```python
#fire_lulc_df['acq_date'] = pd.to_datetime(fire_lulc_df['acq_date'], format='%Y%m%d %I:%M%p)
fire_lulc_df['month'] = pd.DatetimeIndex(fire_lulc_df['acq_date']).month
```

```python
fire_lulc_df['period_day'] = pd.cut(fire_lulc_df['acq_time'], bins=[0, 500, 1200, 1700, 2100, 2400], include_lowest=True, labels=['night', 'morning', 'afternoon', 'evening', 'night'], ordered=False)
```

```python
#boolean field representing either period of planting and harvesting harvesting/planting = Yes/1 and No/0

def getagricPeriod(month):
    if (month == 4 or month == 5 or month == 6 or month == 7):
       return 'Yes'
    elif(month == 9 or month == 10 or month == 11 or month == 12):
       return 'Yes'
    else:
       return 'No'

fire_lulc_df['plant_harvest'] = fire_lulc_df['month'].apply(getagricPeriod)
```

```python
fire_lulc_df
```

```python
# month of the year
fire_lulc_df['month'] = fire_lulc_df['month'].replace(to_replace=1, 
                                                        value="January")
fire_lulc_df['month'] = fire_lulc_df['month'].replace(to_replace=
                                                        2,
                                                        value="February")
fire_lulc_df['month'] = fire_lulc_df['month'].replace(to_replace=3, 
                                                value="March")
fire_lulc_df['month'] = fire_lulc_df['month'].replace(to_replace=4, 
                                                        value="April")
fire_lulc_df['month'] = fire_lulc_df['month'].replace(to_replace=5, 
                                                        value="May")
fire_lulc_df['month'] = fire_lulc_df['month'].replace(to_replace=6, 
                                                        value="June")
fire_lulc_df['month'] = fire_lulc_df['month'].replace(to_replace=7, 
                                                        value="July")
fire_lulc_df['month'] = fire_lulc_df['month'].replace(to_replace=8, 
                                                        value="August")
fire_lulc_df['month'] = fire_lulc_df['month'].replace(to_replace=9, 
                                                        value="September")
fire_lulc_df['month'] = fire_lulc_df['month'].replace(to_replace=10, 
                                                        value="October")
fire_lulc_df['month'] = fire_lulc_df['month'].replace(to_replace=11, 
                                                        value="November")
fire_lulc_df['month'] = fire_lulc_df['month'].replace(to_replace=12, 
                                                        value="December")
```

```python
fire_lulc_df
```

```python
fire_lulc_df1 = fire_lulc_df.groupby('month').size().reset_index(name ='fireIncidents')
fire_lulc_df1 = fire_lulc_df1.sort_values(by = 'fireIncidents', ascending= False)
fire_lulc_df1['ps_ratio'] = fire_lulc_df1['fireIncidents']/fire_lulc_df1['fireIncidents'].sum() * 100
fire_lulc_df1
```

```python
plt.figure(figsize= (15,10))
plt.title("Fire Incident by Month from the year 2015-2021")
sns.barplot(data = fire_lulc_df1, x = "month", y = "fireIncidents")
plt.xlabel("months")
```

We detect a spike in instances throughout specific months of the year. March had almost 6000+Â fires, accounting for ~35% of all fires. February, with over 4000+Â instances, was the following month. They account for 57% of all reported cases. According to research, these two months are almost never specified as planting or harvesting seasons for the majority of crops; thus, they correlate to dates when neither planting nor harvesting occurs. Also note that the first 4 months of the year have high fires occurrence, which also correlate with expert insight on farming practices in telangana. 

However, some crops planting/harvesting are done during January, April, and May, which varies from villages to villages but may be the reason for lower fire cases during these months as compared to February and March (culturally and by norm are generally wait periods; hence, nothing happens these months).

They are a somewhat relaxed period from planting and harvesting and engage in Slash-and-burn agriculture (shifting cultivation)Â in Telangana and other neighboring divisions. This period (February and March) for now can be hypothesisedÂ as caused by agricultural practices as they are thought of as a form of agroforestry and done before the rainiest part of the year to start a new agricultural cycle.

```python
plt.figure(figsize= (12,8))
plt.title('General Period for Planting and Harvesting')
sns.histplot(data = fire_lulc_df, x= fire_lulc_df.plant_harvest, hue =fire_lulc_df.plant_harvest)
plt.xlabel('Planting/harvesting')
plt.ylabel('Fires Incidents')
```

From the above, it can be seen that most fire incident occurs during the period where there is generally either no planting or harvesting activities in Telangana.

```python
firemonlulc = fire_lulc_df.groupby(['month', 'lulc']).size()
firemonlulc = firemonlulc.rename_axis(['month', 'landuse'])\
                               .unstack('landuse')\
#                               .reindex(columns=toas)
plt.figure(figsize=(15,10))
plt.style.use('dark_background')
sns.heatmap(firemonlulc, cmap='Spectral')
plt.title('\nFire Cases by Month per Land Use\n', fontsize=14, fontweight='bold')
plt.xticks(fontsize=15)
plt.yticks(fontsize=12)
plt.xlabel('')
plt.ylabel('')
plt.savefig('firebymonthperlanduse.png')
plt.show()
```

From further multivariate analysis, it can be seen that built-up area represent has a lesser number of fire incident for the Month of March and February. 'Moss and lichen' and the 'Shrubland' land use, which are areas containing fertile soil seems to have the highest incident of fire for the two month with highest fire incidents. This could definitely signal that for this months in particular, fire are majorly caused by agricultural practices. This correlates with the fact that most farmers generally use these periods for shifting cultivation (Slash-and-burn agriculture).

```python
#normalize the numerical(integer/continuous) features in our data to ensure that one feature is not interpreted as being more important than the other.
from sklearn import preprocessing
from kmodes.kprototypes import KPrototypes

fireIncident = fire_lulc_df.copy()
scaler = preprocessing.MinMaxScaler()
fireIncident[['brightness','confidence', 'bright_t31', 'frp']] = scaler.fit_transform(fireIncident[['brightness','confidence', 'bright_t31', 'frp']])
```

```python
#convert fire_lulc DataFrame to numpy array
fireInc_array = fireIncident.values
```

```python
fireInc_array[:, 0]  = fireInc_array[:, 0].astype(float)
fireInc_array[:, 1]  = fireInc_array[:, 1].astype(float)
fireInc_array[:, 2]  = fireInc_array[:, 2].astype(float)
fireInc_array[:, 3]  = fireInc_array[:, 3].astype(float)
```

```python
fireInc_array
```

```python
kproto = KPrototypes(n_clusters=2, verbose=2, max_iter=85)
clusters = kproto.fit_predict(fireInc_array, categorical=[4, 5, 6, 7, 8]) #4, 5, 6, 7, 8
```

```python
print(kproto.cluster_centroids_)
```

```python
cluster_dict = []
for c in clusters:
    cluster_dict.append(c)
```

```python
cluster_dict
```

```python
fid = fire_lulc['FID_1']
fid
```

```python
fire_lulc_df['fireID'] = fid
fire_lulc_df['class'] = cluster_dict
fire_lulc_df.drop(['brightness','confidence', 'bright_t31', 'frp', 'daynight', 'lulc', 'month', 'period_day', 'plant_harvest'], axis=1, inplace=True)
```
