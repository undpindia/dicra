```python
# loading fires dataset
datasets = "/home/sherwood/learning/Data4Policy/Geospatial Data Science Internship/Datasets/"


telangana_fires = gpd.read_file(datasets + 'telangana_fires.csv')

telangana_fires = gpd.GeoDataFrame(telangana_fires,
                                   geometry = gpd.points_from_xy(
                                       telangana_fires.longitude,
                                       telangana_fires.latitude),
                                    crs = 'EPSG:4326'
                                   )

telangana_fires
```

```python
# loading districts
shp_file = "/mnt/c/Users/amoam/Downloads/gadm40_IND_shp/"

telangana_districts = gpd.read_file(shp_file + 'gadm40_IND_3.shp')

telangana_districts = telangana_districts[telangana_districts['NAME_1'] == 'Telangana']

telangana_districts

```

While exploring the dataframe, two district names are `n.a. ( 1711)` and `n.a. ( 1728)` which suggests more research has to be done to change their name. However, since these are the <b>_distric name_</b> available from [GADM Level 3](https://geodata.ucdavis.edu/gadm/gadm4.0/kmz/gadm40_IND_3.kmz), I will leave them as they are.

```python
# spatial join
telangana_fires_sj = gpd.sjoin(telangana_fires,telangana_districts, op = 'within')
```

```python
# extract first 4 elements from that col
telangana_fires_sj['year'] = telangana_fires_sj['acq_date'].str[:4].astype(int)
```

```python
telangana_fires_sj
```

```python
output = telangana_fires_sj[['NAME_3','year','instrument']].groupby(['NAME_3','year']).count()
output = output.reset_index()
output = output.set_axis(['district','year','fireCount'],axis=1)
output
```
