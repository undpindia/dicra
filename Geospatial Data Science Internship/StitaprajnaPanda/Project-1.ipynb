```python

        ##################################      PART-1      ###########################################


import pandas as pd
import numpy as np
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import tqdm
from tqdm._tqdm_notebook import tqdm_notebook
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\laptop\Desktop\DataScience\telangana_fires.csv', parse_dates=['acq_date'])
df.head()
```

```python
df['year'] = pd.DatetimeIndex(df['acq_date']).year
```

```python
df.head()
```

```python
df1 = df.copy()
df1.drop('acq_date',inplace=True,axis=1)
```

```python
df1.head()
```

```python
df1['Adrs'] = df1['latitude'].map(str) + ' , ' + df1['longitude'].map(str)
```

```python
locator = Nominatim(user_agent='myGeocoder',timeout=10)
rgecode = RateLimiter(locator.reverse, min_delay_seconds=0.001)
```

```python
# This converts coordinates to physicsl address
tqdm_notebook.pandas()
df1['Address'] = df1['Adrs'].progress_apply(locator.geocode)
```

```python
df1.isnull().sum() # Checking if there is any null value
```

```python
df1.head(6)
```

```python
df1.drop(['latitude','longitude','version','instrument','Adrs'],axis=1,inplace=True)
```

```python
df1.tail(6)
```

```python
df1.groupby('type')['type'].agg('count')
```

```python
df1.groupby('daynight')['daynight'].agg('count')
```

```python
df1.to_csv('New_Telangana_fire.csv')
```
