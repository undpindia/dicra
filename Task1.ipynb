```python
#Calling Libraries
import pandas as pd
import geopandas as gpd
import matplotlib as plt
from osgeo import gdal as gd
```

```python
#reading and visualising data
data=pd.read_csv('/content/drive/MyDrive/telangana_fires.csv')
data.info()
vector = gpd.read_file('/content/drive/MyDrive/telangana.shp')
print(vector)
vector.head()
```

```python
#conversion of csv to point data
points = gpd.GeoDataFrame(data, geometry=gpd.points_from_xy(data.longitude, data.latitude),crs=4326)
points.info
```

```python
#Spatialjoin
import rtree
fire =gpd.sjoin(points,vector,how="right",predicate="within")
fire.head()
```

```python
#Selection of required data
selected=fire[['NAME_3', 'year','geometry']]
selected.head()
```

```python
#Counting the fire instances per administrative boundary per year
from pandas.core.frame import DataFrame
a = selected.groupby(['NAME_3', 'year']).size().reset_index(name='count')
count1=pd.DataFrame(a)
df.columns = ['NAME_3','year','count']
df.head()

```

```python
#Merging count data with selected data
merged_areas = selected.merge(a, on='NAME_3', how='outer')
merged_areas.head()
```

```python
merged_areas.plot(column="count", cmap="Red_r", figsize=(12,10));
```

```python
dataFrame = pd.DataFrame(merged_areas, columns = ['NAME_3','year','count'])
json = dataFrame.to_json(df.to_json(r'/content/drive/MyDrive/firecount.json'))
```
