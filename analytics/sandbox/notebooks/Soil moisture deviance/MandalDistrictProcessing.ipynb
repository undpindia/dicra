To obtain, district and mandal level data, We have used **rasterToVector.py** script 

```python
district_path = "convert\\districts"
mandal_path = "convert\\mandals"
```

```python
district_dataset = [ds for ds in os.listdir(district_path) if ds[-8:] == '.geojson']
mandal_dataset = [ds for ds in os.listdir(mandal_path) if ds[-8:] == '.geojson']
```

```python
len(district_dataset)
```

```python
mandal = gpd.read_file(mandal_path+"/"+mandal_dataset[0])
district =  gpd.read_file(district_path + "/" + district_dataset[0])
```

```python
ssm = []
for i in range (0, len(mandal)):
    ssm.append(mandal["zonalstat"][i]['mean'])
```

```python
mandal["ssm"] = ssm
```

```python
mandal = mandal[["Mandal_Nam","geometry","ssm"]]
```

```python
ssm = []
for i in range (0, len(district)):
    ssm.append(district["zonalstat"][i]['mean'])
```

```python
district["ssm"] = ssm
```

```python
district
```

```python
district = district[["Dist_Name","geometry","ssm"]]
```
